```python
import re

qualifications = ["diploma", "higher national diploma", "bachelor's degree", "master's degree",
                  "doctorate degree", "vocational training institute" , "college of education",
                 "high school certificate"]


pattern = '|'.join(qualifications)

new_df['Job Description'] = new_df['Job Description'].astype(str)


def extract_qualification(text):
    matches = re.findall(pattern, text)
    return ', '.join(matches) if matches else None

new_df['Qualification'] = new_df['Job Description'].apply(extract_qualification)
```

```python
pattern = r'\b(\d{1,2})\s*[-/yY]{1,2}\s*(\d{1,2})\s*[-/yY]{1,2}'

# Define a function to extract the years of experience
def extract_experience(text):
    matches = re.findall(pattern, text)
    if matches:
        start, end = matches[0]
        return (int(end) - int(start) + 1)
    else:
        return None

# Extract the years of experience into a new column 'Experience'
new_df['Experience'] = new_df['Job Description'].apply(extract_experience)
```

```python
# imports
import pandas as pd
import spacy
import en_core_web_lg
from spacy.matcher import PhraseMatcher

# import skill extractor
from skillNer.skill_extractor_class import SkillExtractor
from skillNer.general_params import SKILL_DB


# init params of skill extractor
nlp = en_core_web_lg.load()
# init skill extractor
skill_extractor = SkillExtractor(nlp, SKILL_DB, PhraseMatcher)
```

```python
import neattext.functions as nfx
from neattext.functions import clean_text
from neattext import TextCleaner
```

```python
def clean(text):
  docx = TextCleaner(text.lower())
  result = docx.remove_emails().remove_urls().remove_dates().remove_html_tags().remove_numbers().remove_puncts().remove_stopwords().remove_special_characters()
  doc = nlp(result.text)
  empty_list = []
  for token in doc:
    empty_list.append(token.lemma_)
  final_string = ' '.join(map(str,empty_list))
  return final_string
```

```python
def get_skill(text):
  try:
    annotations = skill_extractor.annotate(text,tresh=0.5)
    skill_dict={"Soft Skill":[],"Hard Skill":[]}
    soft_skill = []
    hard_skill = []

    for item in annotations['results']['ngram_scored']:
      skill_id = item['skill_id']
      skill_type = skill_extractor.skills_db[skill_id]['skill_type']
      
      if skill_type == 'Soft Skill' and item['doc_node_value']:
        soft_skill.append(item['doc_node_value'])
      if skill_type == 'Hard Skill':
        hard_skill.append(item['doc_node_value'])  
    skill_dict['Soft Skill'] =set(soft_skill)
    skill_dict['Hard Skill'] = set(hard_skill)
    return(skill_dict)
  except:
    return None
```
